```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

from collections import deque
class Solution:

    com = None
    def findParent(self, node, p, q):
        if (not node):
            return False
        
        mid = (node.val == p.val or node.val==q.val)
        left = self.findParent(node.left, p, q)
        right = self.findParent(node.right, p, q)

        if (mid + left + right) >= 2:
            self.com = node
            return True
        
        return mid or left or right
                        
    # find path is very slow
    def findPath(self, node, tar):
        if (not node):
            return []
        
        if (node.val == tar.val):
            return [node]
        
        l = self.findPath(node.left, tar)
        if (l):
            return [node] + l
        
        r = self.findPath(node.right, tar)
        if (r):
            return [node] + r
        
        return []
        
    
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        
        if (not root):
            return None

        '''
        ppath = self.findPath(root, p)
        qpath = self.findPath(root, q)

        p = None
        for i in range(min(len(ppath), len(qpath))):
            if (ppath[i] == qpath[i]):
                p = ppath[i]
            else:
                break
        return p
        '''
        self.findParent(root, p, q)
        return self.com
                
            
```
