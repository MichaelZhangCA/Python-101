```python
class TrieNode():
    
    def __init__(self):
        self.children = {}
        self.end = False
        
        
class Trie:
    
    def __init__(self):
        self.root = TrieNode()
        
    def insert(self, word):
        
        cur = self.root
        for c in word:
            if (c not in cur.children):
                cur.children[c] = TrieNode()
            
            # move to next layer
            cur = cur.children[c]
            
        cur.end = True

    # by given any prefix, return all possible words
    def search(self, pre):
        
        words = []
        cur = self.root
        for c in pre:
            if (c not in cur.children):
                return []
            
            cur = cur.children[c]
        
        if (cur.end):
            words.append(pre)

        # continue to all children combinations
        words += self._iterate_children(cur, pre)
        return words

    def _iterate_children(self, node, word):
        words = []
        for k, v in node.children.items():
            v = node.children[k]
            if v.end:
                words.append(word+k)
        
            words += self._iterate_children(v, word+k)
        
        return words
```
